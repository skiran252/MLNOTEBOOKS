```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
import numpy as np 
```

```python
data=pd.read_csv('play_tennis.csv')
data.head()
```

```python
X=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
data.describe()
```

```python
one_hot_data = pd.get_dummies(data[['outlook','temp','humidity','wind']])
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(one_hot_data,data['play']) 
```

```python
from sklearn import tree
import os
os.environ['PATH'] = os.environ['PATH']+';'+os.environ['CONDA_PREFIX']+r"\Library\bin\graphviz"
```

```python
import pydotplus
from IPython.display import Image
dot_data = tree.export_graphviz(model, out_file=None, feature_names=list(one_hot_data.columns.values), 
                              class_names=['Not Play', 'Play'], rounded=True, filled=True)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
model.score(one_hot_data,data['play'])
```
