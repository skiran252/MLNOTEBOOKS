### 3) Print the common items from two Numpy arrays and also print the position of the common elements and remove the common items and print the different elements also of two   Numpy arrays?

```python
a1=np.array([1,2,5,6])
a2=np.array([1,2,6,7])
for i in a1:
    if i in a2:
        print("common element in both arrays is",i)
        i1=list(a1).index(i)
        i2=list(a2).index(i)
        print("pos of element in array 1:",i1+1,"\npos of element in array 2:",i2+1)
        temp=list(a1)
        temp.remove(i)
        a1=np.array(temp)
        print("The array 1 after removing %d is " % i, a1)
        temp=list(a2)
        temp.remove(i)
        a2=np.array(temp)
        print("The array 1 after removing %d is " % i, a2)
        
```

```python
a1=np.array([1,2,5,6])
a2=np.array([1,2,6,7])
c=np.intersect1d(a1,a2,return_indices=True)
print(c)
a1=np.delete(a1,c[1])
a2=np.delete(a2,c[2])
print(a1)
print(a2)
```

### 8)Create a line plot for the given equation y=x^3 and plot some points using linspace() in the range (-10,10) and also label x-axis, y-axis and give the title of graph cubic function and display the graph


```python
x=np.linspace(-10,10)
x
y=x**3
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y=x^3")
plt.title("Cubic Function")
plt.show()
```

### 10)Represent two equations y=x^3&y=x^2 in one graph and useLegend function.

```python
x = np.linspace(0, 10)
y1 = x**3
y2 = x**2

plt.plot(x,y1,x,y2)
plt.legend(('y^3','y^2'))
plt.title("USING LEGEND")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
